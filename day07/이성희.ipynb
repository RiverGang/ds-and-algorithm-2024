```python
# 스택 풀 확인함수
def 스택풀확인():
    global SIZE, stack, top
    if top == (SIZE -1):
        return True # 스택이 꽉 찼음
    else:
        return False
    
# 스택 데이터 삽입함수
def 스택삽입(data):
    global SIZE, stack, top
    if 스택풀확인() == True:
        print('스택이 꽉 찼습니다')
        return
    else:
        top += 1
        stack[top] = data

# 스택 엠티 확인함수
def 스택빈공간확인():
    global SIZE, stack, top
    if top <= -1:
        return True # 스택이 비었음
    else:
        return False
    
# 스택 데이터 추출함수
def 스택추출():
    global SIZE, stack, top
    if 스택빈공간확인() == True:
        print('스택이 비었습니다')
        return None
    else:
        data = stack[top]
        stack[top] = None
        top -= 1
        return data

# 스택 최종데이터 확인함수
def 스택피크확인():
    global stack, top
    if 스택빈공간확인() == True:
        print('스택이 비어있습니다.')
        return None
    else:
        return stack[top]
    

# 전역변수 선언
SIZE = int(input('스택 크기 입력 -->'))
stack = [None for _ in range(SIZE)]
top = -1

# 메인코드
if __name__ == '__main__':
    select = input('삽입(i), 추출(e), 확인(v), 종료(x) >> )')

    while (select.lower()) != 'x':
        if select.lower() == 'i':
            data = input('입력데이터 >> ')
            스택삽입(data)
            print(f'스택상태 : {stack}')

        elif select.lower() == 'e':
            data = 스택추출()
            print(f'추출 데이터 : {data}')
            print(f'스택상태 : {stack}')

        elif select.lower() == 'v':
            data = 스택피크확인()
            print(f'확인 데이터 : {data}')
            print(f'스택상태 : {stack}')

        else:
            print('입력이 잘못됨')
    
        select = input('삽입(i), 추출(e), 확인(v), 종료(x) >> )')
    
    print('프로그램 종료')
    
```

```python
# Queue 풀확인함수
def isQueueFull():
    global SIZE, queue, front, rear
    if rear != (SIZE -1): 
        return False
    elif rear == (SIZE-1) and front == -1: 
        return True
      

# Queue 엠티함수
def isQueueEmpty():
    global front, rear
    if front == rear:
        return True
    else:
        return False
    
# Queue 데이터삽입함수
def enQueue(data):
    global queue, rear
    if isQueueFull() == True:
        print ('대기줄이 모두 찼습니다')
        return
    else:
        rear += 1
        queue[rear] = data

# Queue 데이터추출함수
def deQueue():
    global queue, front, rear
    if isQueueEmpty() == True:
        print('큐가 비었습니다')
        return
    else:
        front += 1
        data = queue[front]
        queue[front] = None

        for i in range(front+1, rear+1):
            queue[i-1] = queue[i]
            queue[i] = None
        front = -1
        rear -= 1

        return data
        
# 전역변수
SIZE = int(input('대기줄 크기 입력(정수) > '))
queue = [None for _ in range(SIZE)]
front = -1 
rear = -1

if __name__ == '__main__': #메인 시작
    while True:
        select = input('대기 등록(e), 대기 입장(i), 종료(x) > ')
       
        if select.lower() == 'e':
           data = input('입력 데이터 > ')
           enQueue(data)
           print(f'대기줄 상태 : {queue}')
        
        elif select.lower() == 'i':
            data = deQueue()
            print(f'"{data}"님 식당에 들어감')
            print(f'대기줄 상태 : {queue}')
    
        elif select.lower() == 'x':
            break
        else:
            continue
    
    print("식당 영업 종료!")
```
